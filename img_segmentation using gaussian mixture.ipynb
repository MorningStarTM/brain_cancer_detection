```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
img = cv2.imread("E:\ML_test\Data\zipped\Copy of Tr-pi_0010.jpg")
img.shape
```

```python
#reshape the array (512x512)
img2 = img.reshape((-1, 3))
img2.shape
```

```python
#segment is 3
gmm = GaussianMixture(n_components=3, covariance_type="tied").fit(img2)
```

```python
gmm_labels = gmm.predict(img2)
```

```python
#reshaping array again into image format
segmented_img = gmm_labels.reshape(img.shape[0], img.shape[1])
```

```python
fig, ax = plt.subplots(1,2, figsize=(18,9))
ax[0].imshow(img)
ax[1].imshow(segmented_img)
```
